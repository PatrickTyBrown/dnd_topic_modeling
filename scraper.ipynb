```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
import regex as re
```

```python
html = requests.get('https://criticalrole.fandom.com/wiki/Category:Transcripts').text
soup = BeautifulSoup(html)
# print(soup)
```

```python
tags = soup.findAll('li', {'class':'category-page__member'})
for tag in tags:
    tag = tag.find('a')
    if tag.text and tag['href'] and bool(re.search('Transcript', tag.text)):
        print(tag['href'])
```
